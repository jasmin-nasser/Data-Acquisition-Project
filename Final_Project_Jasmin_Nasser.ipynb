```python
import matplotlib.pyplot as plt
import arabic_reshaper
from bidi.algorithm import get_display
import networkx as nx

G = nx.DiGraph()


for result in search_data:
    title = get_display(arabic_reshaper.reshape(result["title"]))
    snippet = get_display(arabic_reshaper.reshape(result["snippet"]))
    G.add_node(title, type="title")
    G.add_node(snippet, type="snippet")

for result in search_data:
    title = get_display(arabic_reshaper.reshape(result["title"]))
    snippet = get_display(arabic_reshaper.reshape(result["snippet"]))
    G.add_edge(title, snippet)

main_node = get_display(arabic_reshaper.reshape("التريند"))
G.add_node(main_node, type="main")

for title_node in G.nodes():
    if G.nodes[title_node]["type"] == "title":
        G.add_edge(main_node, title_node)

pos = nx.spring_layout(G, scale=1)

plt.figure(figsize=(30, 17))

nx.draw_networkx_nodes(G, pos, node_size=1000)
nx.draw_networkx_edges(G, pos, width=3, alpha=0.5, edge_color='b')
nx.draw_networkx_labels(G, pos, font_size=10, font_family="DejaVu Sans")

plt.axis('off')
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)

betweenness_centrality = nx.betweenness_centrality(G)

print("node\tDegree Centrality\tBetweenness Centrality")

for node in G.nodes():
    first_five_words = " ".join(node.split()[:5])
    degree_cent = degree_centrality.get(node, 0)
    betweenness_cent = betweenness_centrality.get(node, 0)
    print(f"{get_display(arabic_reshaper.reshape(first_five_words))}\t\t{degree_cent:.3f}\t\t\t{betweenness_cent:.3f}")
```

```python
communities = list(nx.algorithms.community.greedy_modularity_communities(G))
print("\nCommunities:")
for i, community in enumerate(communities):
    print(f"Community {i + 1}: {community}")
```
